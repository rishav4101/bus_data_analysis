```python
import pandas as pd
data=pd.read_csv("gps.csv")
#Preprocessing data
data.columns = ['Latitude', 'Longitude', 'Speed', 'Altitude', 'TimeStamp', 'stop/busy']
data.drop(['stop/busy'], axis=1, inplace=True)
data['Date']=pd.to_datetime(data['TimeStamp']).dt.date
data['Time']=pd.to_datetime(data['TimeStamp']).dt.time
data.drop(['TimeStamp'], axis=1, inplace=True)

def Daytime(Time):
    hr=int(Time.hour)
    if hr<9:
        return 'early morning'
    elif 9<=hr<12:
        return 'morning'
    elif 12<=hr<17:
        return 'afternoon'
    elif 17<=hr:
        return 'evening'
    
data['Daytime']=data.Time.apply(Daytime)
data1=data
data1

df=pd.read_csv("processed_Bus_SOUND_2019_09_23_18_05_19_187_0.txt", sep=',')
df.to_csv("sound.csv")
data2=pd.read_csv("sound.csv")
data2.columns = ['lindex', 'Date', 'Time', 'Intensity']
data1
```
