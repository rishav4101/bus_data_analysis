```python
import pandas as pd

#reading file
#provide GPS file address
df=pd.read_csv("Prantika-DGP 38/All/Bus_GPS_2019_09_23_18_05_21_609.txt", sep=',')

df.to_csv("gps.csv")

data=pd.read_csv("gps.csv")

```

```python
#Preprocessing data
data.columns = ['Latitude', 'Longitude', 'Speed', 'Altitude', 'TimeStamp', 'stop/busy']
data.drop(['stop/busy'], axis=1, inplace=True)
data['Date']=pd.to_datetime(data['TimeStamp']).dt.date
data['Time']=pd.to_datetime(data['TimeStamp']).dt.time
data.drop(['TimeStamp'], axis=1, inplace=True)

def Daytime(Time):
    hr=int(Time.hour)
    if hr<9:
        return 'early morning'
    elif 9<=hr<12:
        return 'morning'
    elif 12<=hr<17:
        return 'afternoon'
    elif 17<=hr:
        return 'evening'
    
data['Daytime']=data.Time.apply(Daytime)
fdata=data
fdata
```

```python
#Algo 1 to get stop points of the bus
a1 = fdata[(fdata['Speed']*3.6 <= 0.1)].reset_index()
a1.drop(['Altitude','Date','Time','Daytime'],axis=1,inplace=True)
a1['Distance']=0

from math import sin, cos, sqrt, atan2, radians

import math as Math
def dist(lat1,lon1,lat2,lon2):
    R = 6371e3 #meters
    Ï†1 = Math.radians(lat1)
    Ï†2 = Math.radians(lat2)
    Î”Ï† = Math.radians(lat2-lat1)
    Î”Î» = Math.radians(lon2-lon1)

    a = Math.sin(Î”Ï†/2) * Math.sin(Î”Ï†/2) +\
            Math.cos(Ï†1) * Math.cos(Ï†2) *\
            Math.sin(Î”Î»/2) * Math.sin(Î”Î»/2)
    c = 2 * Math.atan2(Math.sqrt(a), Math.sqrt(1-a))

    d = R * c
    return d

l=a1.index.values
for i in range(2,len(l)):
    a1.loc[l[i],'Distance']=dist(a1.loc[l[i],'Latitude'],a1.loc[l[i],'Longitude'],a1.loc[l[i-1],'Latitude'],a1.loc[l[i-1],'Longitude'])

final_stops = a1[(a1['Distance']>=5.00)].reset_index()
final_stops.drop(['level_0','Distance','index'],axis=1,inplace=True)
final_stops
        
#Nicee We get all 26 Stop points.
```
