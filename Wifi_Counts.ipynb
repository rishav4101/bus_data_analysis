```python
df = data.drop(['Unnamed: 0'], axis=1)
df.columns = ['Wifi_add', 'Wifi_name', 'strength', 'Timestamp']
df
```

```python
df['Timestamp']=df['Timestamp'].astype(str)
data_gps['Timestamp']=data_gps['Timestamp'].astype(str)
data_gps=data_gps.merge(df, on='Timestamp',how="left")
data_gps=data_gps.dropna()
data_gps
```

```python
data_gps['count']=0

from math import sin, cos, sqrt, atan2, radians

import math as Math
def dist(lat1,lon1,lat2,lon2):
    R = 6371e3 #meters
    φ1 = Math.radians(lat1)
    φ2 = Math.radians(lat2)
    Δφ = Math.radians(lat2-lat1)
    Δλ = Math.radians(lon2-lon1)

    a = Math.sin(Δφ/2) * Math.sin(Δφ/2) +\
            Math.cos(φ1) * Math.cos(φ2) *\
            Math.sin(Δλ/2) * Math.sin(Δλ/2)
    c = 2 * Math.atan2(Math.sqrt(a), Math.sqrt(1-a))

    d = R * c
    return d

l=data_gps.index.values
start=0
for i in range(1,len(l)):
    if((dist(data_gps.loc[l[i],'Latitude'],data_gps.loc[l[i],'Longitude'],data_gps.loc[l[start],'Latitude'],data_gps.loc[l[start],'Longitude']))>=200):
        WifiC = data_gps["Wifi_add"][start:i].nunique()
        data_gps.loc[l[i],'count']=WifiC
        start=i

data_gps=data_gps[(data_gps['count']>0)]
data_gps.drop(['Altitude','Timestamp','Speed','Wifi_add','Wifi_name','strength'],axis=1,inplace=True)
data_gps       
        
```
